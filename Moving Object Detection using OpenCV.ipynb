# Display the camera until using the terminate key

```python
import cv2
import time

cam = cv2.VideoCapture(0) # Initialize the camera with the primary camera
time.sleep(1) # How much time it will take 

while True:
    _,img = cam.read() # Read the image
    cv2.imshow("CameraFeed", img) # Display the image from the camera.
    
# If use imshow() function,  initialize waitKey() and destroyAllWindows() function
    key = cv2.waitKey(1) & 0xFF  #Generate a key  # Return value based on the key press, 0xFF means return 8 bit 1s
    if key== ord("q"):   # If press q then terminate
        
        break
        
cam.release()   # Release the camera
cv2.destroyAllWindows()   # Close all window
```

# Full program - Moving object detection using OpenCV

## Steps:
- Feed from the camera
- Have to capture the first image to fix it as a backbround image
- Make a comparision between the background image and the current frame

```python
import cv2
import time
import imutils  # for resizing the program

cam = cv2.VideoCapture(0)
time.sleep(1)

# Deploy resize to have a better view
# Initializing for background

FirstFrame = None
area = 500


while True:
    _,img = cam.read()
    
    text = "Normal"
    
    img = imutils.resize(img, width = 500) # Resize the image
    grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Gray scale image
    gaussianImg = cv2.GaussianBlur(grayimg, (21,21), 0) # Gaussian Image--> (21,21)-->kernel
    
    # Taking the first frame for background
    
    if FirstFrame is None:
        FirstFrame = gaussianImg
        continue
    
    imgdiff = cv2.absdiff(FirstFrame, gaussianImg) #Compare the background with the current frame
    
    threshImg = cv2.threshold(imgdiff, 25, 255, cv2.THRESH_BINARY)[1] # Taking the threshold image
    threshImg = cv2.dilate(threshImg, None, iterations = 2) # Holes can be removed from threshold image
    
    # By using contour we can plot the rectangle whereever the object is in movement
    cnts = cv2.findContours(threshImg.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts) # Taking each and every contours
    
    #Passing each contour to detect whether it is in movement or not.
    
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
            
        (x, y, w, h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        text = "Moving object detected"
    
    print(text)
   
    
    # Putting text into image
    
    cv2.putText(img, text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    
    #For display
    
    cv2.imshow("CameraFeed", img)
    
    key = cv2.waitKey(1) & 0xFF
    
    if key==ord("q"):
        break
        
cam.release()  # releasing the camera
cv2.destroyAllWindows() # destroy all windows
```
